```python
import torch
from torchvision import transforms
import os
import re
from PIL import Image
import cv2
import numpy as np

from models.inception_resnet_v1 import InceptionResNetV1
```

```python
model = InceptionResNetV1(pretrained='vggface2')
```

```python
baseline_image_path = '../results/finetune_edges2shoes/test_latest/images'
iterrefinec_image_path = '../results/finetune_iterrefine_e2e/test_latest/images'
iterrefineg_image_path = '../results/finetune_iterrefine_g/test_latest/images'
iterrefinec_2gan_image_path = '../results/finetune_iterrefine_e2e_2gan/test_latest/images'
iterrefinec_1gan_image_path = '../results/finetune_iterrefine_e2e_1gan/test_latest/images'
```

```python
"""
Constructs image path pairs for the output data for the test set with the images inside `dir`.

`dataset` is one of 'baseline', 'iterrefinec', 'iterrefineg'.
`mode` is either 'final' or 'intermediate', denoting either the final output image of the GAN 
or the output image from the intermediate model (net_G rather than net_G_2).
"""
def make_image_path_pairs(dir, dataset='iterrefinec', mode='final'):
    image_pairs = []
    assert os.path.isdir(dir), '%s is not a valid directory' % dir
    assert dataset in ['baseline', 'iterrefinec', 'iterrefineg']
    assert mode in ['final', 'intermediate']

    fake_regex = r'(.*)_fake_B.png'
    real_regex = r'(.*)_real_B.png'
    if dataset != 'baseline' and mode == 'final':
        fake_regex = r'(.*)_fake_B_2.png'
    
    fake_paths = {}
    real_paths = {}

    for root, _, fnames in sorted(os.walk(dir)):
        for fname in fnames:
            path = os.path.join(root, fname)
            match = re.search(fake_regex, fname)
            if match:
                key = match[1]
                if key in real_paths:
                    real_path = real_paths[key]
                    image_pairs.append((path, real_path))
                    del real_paths[key]
                    continue
                fake_paths[key] = path
                continue

            match = re.search(real_regex, fname)
            if match:
                key = match[1]
                if key in fake_paths:
                    fake_path = fake_paths[key]
                    image_pairs.append((fake_path, path))
                    del fake_paths[key]
                    continue
                real_paths[key] = path
    return image_pairs
```

```python
def get_image(image):
    ret = Image.open(image)
    ret = ret.resize((256, 256))
    ret = ret.convert('RGB')
    normalize = transforms.Compose([transforms.ToTensor()])
    ret = normalize(ret)
    return ret

def l2_dist(vec1, vec2):
    return (vec1 - vec2).pow(2).sum().sqrt().item()

def l1_dist(vec1, vec2):
    return (vec1 - vec2).abs().sum().item()
```

```python
def facenet_dist_calc(image_pair):
    fake = image_pair[0]
    real = image_pair[1]
    pair = torch.stack((fake, real)).detach()
    embeddings = model(pair).detach()
    l1 = l1_dist(embeddings[0], embeddings[1])
    l2 = l2_dist(embeddings[0], embeddings[1])
    return (l1, l2)
```

## Baseline Results

```python
image_path_pairs = make_image_path_pairs(baseline_image_path, 'baseline')
image_pairs = [(get_image(fake), get_image(real)) for fake, real in image_path_pairs]
facenet_distances = [facenet_dist_calc(image_pair) for image_pair in image_pairs]
```

```python
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

print("L1 Distance Distribution")
l1_dists = [pair[0] for pair in facenet_distances]
print('mean:', np.mean(l1_dists), 'median:', np.median(l1_dists), 'variance:', np.var(l1_dists))

# plot the l1 distance distribution
plt.hist(l1_dists, 30, facecolor='blue', alpha=0.5)
plt.show()

print("L2 Distance Distribution")
l2_dists = [pair[1] for pair in facenet_distances]
print('mean:', np.mean(l2_dists), 'median:', np.median(l2_dists), 'variance:', np.var(l2_dists))

# plot the l2 distance distribution
plt.hist(l2_dists, 30)
plt.show()
```
